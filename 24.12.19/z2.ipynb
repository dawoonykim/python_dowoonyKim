```python
import pandas as pd
import folium
```

```python
taxi_data=pd.read_csv("./input/택시승차대 현황.csv",encoding="cp949")
taxi_data
```

```python
m=folium.Map(location=[37.547631,126.942463], zoom_start=15)
folium.Marker([37.547631,126.942463]).add_to(m)
taxi_loc=taxi_data.loc[:,["위치명","위도","경도"]]

m = m.get_root()
m.width = 800
m.height = 600
m
```

```python
m=folium.Map(location=[37.547631,126.942463], zoom_start=15)
folium.Marker([37.547631,126.942463],popup="대흥역").add_to(m)
for index, row in taxi_loc.iterrows():
  # print(f"index : {index}\nrow :\n{row}")
  # print(f"row[2] : {row[2]}")
  folium.Marker([row[1],row[2]],popup=row[0],icon=folium.Icon(color="black",icon="fa-solid fa-taxi",prefix="fa-solid")).add_to(m)
  folium.Circle([row[1],row[2]],radius=100,color="red").add_to(m)

m = m.get_root()
m.width = 800
m.height = 600
m
  
```
