```python
import pandas as pd
import folium
```

```python
taxi_data=pd.read_csv("./input/택시승차대 현황.csv",encoding="cp949")
taxi_data
```

```python
m = folium.Map(location=[37.5502, 126.9287], zoom_start=15)
# m.add_child(folium.LatLngPopup())
taxi_loc = taxi_data.loc[:, ["위치명", "위도", "경도"]]
# taxi_loc_lat = taxi_loc.loc[:, "위도"]
# taxi_loc_lng = taxi_loc.loc[:, "경도"]
# taxi_loc_loc = taxi_loc.loc[:, "위치명"]

for lat, lng, loc in zip(taxi_loc["위도"],taxi_loc["경도"],taxi_loc["위치명"]):
    folium.Marker([lat,lng],popup=loc,icon=folium.Icon(color="pink",icon="fa-solid fa-taxi", prefix="fa-solid")).add_to(m)
    folium.CircleMarker([lat,lng],radius=50,color="skyblue").add_to(m)

m=m.get_root()
m.width="50%"
m.height="500px"

m
```
