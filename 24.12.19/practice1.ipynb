```python
import folium
```

```python
m = folium.Map(location=[37.547631, 126.942463], zoom_start=15,
               tiles="cartodbpositron")
folium.Marker([37.547631, 126.942463], popup="팝업", tooltip="툴팁", icon=folium.Icon(
    color="black", icon="fa-solid fa-subway")).add_to(m)

folium.Marker([37.53959008903884, 126.94590610865076], popup="subway", tooltip="마포역", icon=folium.Icon(
    color="lightblue", icon="fa-solid fa-marker", prefix="fa-solid")).add_to(m)


m=m.get_root()
m.width=800
m.height=500

m
```
