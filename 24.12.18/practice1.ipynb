```python
font1={"family":"fantasy","color":"deeppink","weight":"normal","size":"xx-large"}
plt.plot([1,2,3,4],[1,4,9,16])

plt.xlabel("x",labelpad=30,loc="left",fontdict=font1)
plt.ylabel("y",loc="top")
plt.show()
```
