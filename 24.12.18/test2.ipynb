```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
labels = ["Apple", "Banana", "Mango", "Blueberry"]
sizes = [15, 30, 45, 10]
explode = (0, 0.1, 0, 0)
```

```python
wedgeprops = {"width": 1, "edgecolor": "k", "linewidth": 1}

plt.pie(sizes, labels=labels, autopct="%.1f%%", explode=explode, colors=[
        "CornflowerBlue", "orange", "MediumSeaGreen", "red"], wedgeprops=wedgeprops,startangle=90,shadow=True)
plt.show()
```

```python
x = np.linspace(0, 10, 100)

# sin과 cos 함수의 값을 계산
y_sin = np.sin(x)
y_cos = np.cos(x)
y_tan = np.tan(x)
```

```python
plt.title("Sine and Cosine Functions")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)
plt.plot(x, y_sin, label="sin(x)", color="blue")
plt.plot(x, y_cos, label="cos(x)", color="red")
plt.legend()
```
